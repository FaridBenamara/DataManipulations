# Reading the DataSet

```python
DataFrame = pd.read_csv("Train.csv")
DataFrame 
```

# Testing if number of orders == Number of users

```python
Users = DataFrame["User Id" ].shape
Orders  = DataFrame["Order No" ].shape 
Bikes = DataFrame["Vehicle Type"].shape
print("Number of users = " , Users , "Number of orders " , Orders , "Number of bikes = " , Bikes  )
```

```python
Bikes_0 = DataFrame["Vehicle Type"].astype("category")
Personnal_or_business = DataFrame ["Personal or Business"].astype("category")
print(Bikes_0)
print(Personnal_or_business)
```

# Testing if account categories are only "Personnal and business"

```python
item_counts = DataFrame["Personal or Business"].value_counts(normalize=False  )
item_counts
print(item_counts)
```

# Plot Number of Personnal and business 

```python
comparaison_plot = item_counts.plot.bar( figsize = (4,4) )
plt.figure()
comparaison_plot2 = item_counts.plot.pie( figsize = (6,6) )

```
