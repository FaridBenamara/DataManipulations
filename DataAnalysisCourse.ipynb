#Les dataframes
L’objet DataFrame est un tableau à deux dimensions semblable à un fichier CSV ou à une table de base de données relationnelles. 

Un DataFrame se comporte comme un dictionnaire dont les clefs sont les noms des colonnes et les valeurs sont des séries. 

On peut le considérer alors comme une matrice individus-variables où les lignes correspondent à des observations, les colonnes à des attributs décrivant les individus.

# Feature engineering

```python
df
```

```python
sns.heatmap(df.isnull())
```

```python
#Créer une colonne
df["new_column"] = 5
df.head(1)
```

```python
#Supprimer une colonne
df.drop('new_column', axis=1, inplace=True)
df.head(1)
```

```python
#Créer une colonne avec d'autres valeurs
df['Family'] = df['Parch'] + df['SibSp']
df.head(1)
```

```python
#Supprimer les lignes qui ont des valeurs manquantes
df["Age"].fillna(0)
```

```python
#Supprimer les lignes ou l'age n'est pas renseigné
df = df.dropna(subset=['Age'])
df.info()
```

```python
df = df.dropna(subset=['Fare'])
df.info()
```

```python
df = df.dropna(subset=['Embarked'])
df.info()
```

```python
#Parenthèse : les fonctions lambda -> Fonctions anonymes
def add(a, b):
  return a+b
print(add(1, 2))

lambda_add = lambda a, b : a+b
print(lambda_add(1, 2))
```

```python
#Créer une colonne "A une cabine ou non" ==> Appliquer une fonction à une colonne
df['Cabin'] = df['Cabin'].map(lambda x : 0 if pd.isnull(x) else 1)
```

```python
#Remplacer le numero du ticket par son prefixe
df['Ticket'] = df['Ticket'].map(lambda x : x[0] if x[0].isalpha() else 0)
```

```python
sns.heatmap(df.isnull())
```

#Convertir les données en integer

```python
df
```

```python
df.info()
```

```python
#Convertir les strings en float ou int
df['Sex'].replace(('male','female'), (0,1), inplace = True)
df['Embarked'].replace(('S','C','Q'), (0,1,2), inplace = True)
df['Ticket'].replace(('A', 'P', 'S', 0, 'C', 'W', 'L', 'F'), (0,1,2,3,4,5,6,7), inplace = True)
```

```python
#Afficher la moyenne de survie en fonction du sexe sous forme de graphique
sns.countplot(x="Survived", data=df, hue='Sex')
```

```python
## Variables
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(20, 5))
women = df[df['Sex']=='female']
men = df[df['Sex']=='male']

## Graph
# Women
ax = sns.distplot(women[women['Survived']==1].Age, label='survived', ax=axes[0])
ax = sns.distplot(women[women['Survived']==0].Age, label='not survived', ax=axes[0],)
ax.set_title('Female')

#Men
ax = sns.distplot(men[men['Survived']==1].Age, bins=18, label='survived', ax=axes[1], kde=False)
ax = sns.distplot(men[men['Survived']==0].Age, bins=18, label='not survived', ax=axes[1], kde=False)
ax.set_title('Male')
```
